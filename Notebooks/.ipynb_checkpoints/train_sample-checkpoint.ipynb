Build a sample of QSOs to train our reconstruction model on.  Right now, we're following BG92, so stick to the rest-frame wavelength range $\lambda\lambda4300-5700Å$.  To get this wavelength range for SDSS spectra requires $z<0.6$

```python
from astropy.io import fits
import numpy as np
from astropy.table import Table
```

```python
dr7 = Table.read("/Users/Trevor1/Desktop/data/files/catalogs/dr7_bh_May09_2011.fits")
names = [name for name in dr7.colnames if len(dr7[name].shape) <= 1]
dr7 = dr7[names].to_pandas()
dr7.head()
```

```python
z   = dr7["REDSHIFT"].values
pl  = dr7["PLATE"].values
mjd = dr7["MJD"].values
fib = dr7["FIBER"].values
```

```python
train_sample = dr7.iloc[(z<0.6),:7]
```

```python
train_sample.to_csv("../data/trainsample_dr7_zlt0.6.csv")
```
